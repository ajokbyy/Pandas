```python
import pandas as pd
india_weather =pd.DataFrame( {
    "city": ["mumbai", "delhi", "banglore"],
    "temperature":[32, 45, 30],
    "humidity":[80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":["new york", "chicago", "orlando"],
    "temperature" : [21, 14, 35],
    "humidity":[68, 65, 75]
})
us_weather
```

```python
# Now we are going to join this two data frame
df = pd.concat([india_weather, us_weather])
df
```

```python
# we can also change index 
df = pd.concat([india_weather, us_weather], ignore_index= True)
df
```

```python
# we can also make a group inside the whole table and give them a key, so that whenever we call the key we get our table 
df = pd.concat([india_weather, us_weather], keys = ["india", "us"])
df
```

```python
# df.loc["india"]
df.loc["us"]
```

```python
# now we are going to insert/append our one table just on the side of the existing table
temperature_df = pd.DataFrame({
    "city":["mumbai", "delhi", "bangalore"],
    "temperature":[32, 45, 30],
})
temperature_df
    
```

```python
windspeed_df = pd.DataFrame({
    "city":["mumbai", "delhi", "bangalore"],
    "windspeed":[7, 12, 9],
}, index=[0, 1,2])
windspeed_df
```

```python
# this will do the normal Concat
df = pd.concat([temperature_df, windspeed_df])
df
```

```python
# here it come side by side 
df = pd.concat([temperature_df, windspeed_df], axis = 1)
df
```

```python
# But the data is not always in order
# what if -
# now we are going to insert/append our one table just on the side of the existing table
temperature_df1 = pd.DataFrame({
    "city":["delhi", "mumbai"],
    "temperature":[32, 45],
},index=[1, 0])
temperature_df1
    
```

```python
df = pd.concat([temperature_df1, windspeed_df], axis = 1)
df
```

```python
# we use a funtion know as index which will help us to arrage the data in order
```

```python
# we can also incert on col 
s = pd.Series(["Humid", "Dry", "Rain"], name = "event")
s
```

```python
df = pd.concat([temperature_df, s], axis=1)
df
```
