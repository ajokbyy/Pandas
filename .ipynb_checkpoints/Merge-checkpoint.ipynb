```python
import pandas as pd
df1 = pd.DataFrame({
    "city":["new york", "chicago", "orlando"],
    "temperature":[21, 31, 35]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["chicago", "new york", "orlando", "san francisco"],
    "humidity":[65, 77, 75, 55]
})
df2
```

```python
# now we are going to marge this two data frames using merge functions
df3 = pd.merge(df1, df2, on="city")
df3
# similar to inner SQL join 
```

```python
#outer join - it will give all the data 
df3 = pd.merge(df1, df2, on="city", how="outer", indicator = True)
df3
```

```python
# now we going to see if two table have same title them what merge funtion do
```

```python
df1 = pd.DataFrame({
    "city":["new york", "chicago", "orlando", "baltimore"],
    "temperature":[21, 14, 35, 38],
    "humidity":[65, 68, 71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["chicago", "new york", "san diego"],
    "temperature":[ 14, 35, 38],
    "humidity":[68, 71, 75]
})
df2
```

```python
# here automatically our merge functions give title
df3 = pd.merge(df1, df2, on="city", how = "outer")
df3
```

```python
# we can also change the title
df3 = pd.merge(df1, df2, on="city", how = "outer", suffixes=("_left", "_right"))
df3
```
