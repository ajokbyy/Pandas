```python
import sqlalchemy
import pandas as pd
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/application')
```

```python
df = pd.read_sql_table("customers", engine)
df
```

```python
# we can read one column also
df = pd.read_sql_table("customers", engine, columns=['name'])
df
```

```python
df = pd.read_sql_query("select id,name from customers",engine)
df
```

```python
# here we have written a sql query and run it in pandas
query ='''
SELECT customers.name, customers.phone_number, orders.name, orders.amount
FROM  customers INNER JOIN orders
ON customers.id = orders.customer_id
'''
df = pd.read_sql_query(query, engine)
df
```

```python
# here we are going to load some excel data into our dataframes
df = pd.read_csv("customers.csv")
df
```

```python
# here now we are going to chage the col name so that it will match with our database name 


# here we are going to copy our excel file into sql dataset
df.rename(columns={
    'Customer Name': 'name',
    'Customer Phone': 'phone_number'
}, inplace = True)
df
```

```python
df = df.drop(columns=["id"])
```

```python
df.to_sql(
    name = 'customers',
    con = engine,
    index = False,
    if_exists = 'append'
)
```
