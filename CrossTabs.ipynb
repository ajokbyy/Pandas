```python
# Frequency
#shift + tab - for documentation of any method
# In pandas, crosstab() is used to compute a frequency table (contingency table) that shows how often combinations of categorical values occur.

import pandas as pd
df = pd.read_excel("survey.xls")
df
```

```python
# like in this table right and left is coming aganin and again - we can replace it by numbers
# same goes for male and female  
# we are not changing any thing we are just couting it and making it more easy
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Nationality, df.Sex)
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins = True)
```

```python
# we can make any col x-axic and y-axic
pd.crosstab(df.Nationality, [df.Handedness, df.Sex])
```

```python
pd.crosstab(df.Nationality, [df.Handedness, df.Sex], normalize = "index")
```

```python
import numpy as np
pd.crosstab(df.Sex, [df.Handedness], values = df.Age, aggfunc= np.average)
#Avrage age of female who are left handed
```
