```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data2.csv")
df
```

```python
# we can do replacement(bacis) with this 
new_df = df.replace(-99999, np.nan)
new_df
```

```python
# But this will not give us our desired value 
# we can also use a list 
new_df = df.replace([-99999, -88888], np.nan)
new_df
```

```python
# Replace values based on specific values 
# By using a dictionary
new_df = df.replace({
    'temperature': -99999,
    'windspeed': - 99999,
    'event': '0'
}, np.nan)
new_df
```

```python
new_df = df.replace({
    -99999:np.nan,
    '0':'sunny'
})
new_df
```
