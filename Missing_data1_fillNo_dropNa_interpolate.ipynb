```python
#but our day col is not a string col, it should be data type. So we are going to conver them
df = pd.read_csv("weather_data.csv", parse_dates = ["day"])

#now it is converted to date type
```

```python
# taking our data col as index 
df.set_index('day', inplace = True)
df
```

```python
# Fill Na with 0
new_df = df.fillna(0)
new_df
```

```python
# now converting na value, for every diff col different value by using dictionaly
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
#here in temp and windspeed it will repace na with 0 but in event it will replace it with no event
```

```python
# In the temperature column, there are many 0 values, which can mislead us 
#So now we are going to change it by using forward fill or ffill
#this method is used to copy the value that is present before the na value
# new_df = df.ffill()
# or
new_df = df.bfill()

new_df

#Forward fill (ffill) copies the previous non-NaN value downward.
```

```python
#axis - coping data brows by rows 
new_df = df.bfill(axis = 1)
new_df
```

```python
# if we want the value to repeat only 2 times 
new_df = df.ffill(limit = 1)
new_df
#limit is a parameter that will help in limiting the repetition of a number
```

```python
#interpolate
new_df = df.interpolate()
new_df = df.interpolate(method = "time")
new_df

# It provides a linear interpolation, which gives us a better value
```

```python
#now we are droping all the rows which contain any null values 
new_df = df.dropna()
new_df
```

```python
# If we want to drop a row which contain all the columas na then i use(like all the col value is na excepet index)
new_df = df.dropna(how="all")
new_df

# If you want to drop all those that contain at least 1, 2 or 3 oe etc element, then we use
new_df = df.dropna(thresh = 2)
new_df
```

```python
# so now if we want to see all the datas from 1 to 11 (which is not in our original file) we use 
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
