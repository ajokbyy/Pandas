```python
#pivot allows us to transform or reshape data
# In pandas, we use a pivot table to summarize, reshape, and analyze data by converting raw (long-format) data into a meaningful aggregated table
import pandas as pd
df = pd.read_csv("weather.csv")
df
```

```python
# we can chage our x-axies and our y-axes with the help of pivot
# pivot alway want two parameter index and columns
df.pivot(index="date", columns = "city")
```

```python
# only humidity i want then
df.pivot(index="date", columns = "city", values = "humidity")
```

```python
#pivot table is used to summarize and aggregate data inside dataframe
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="date", columns = "city")
# now this will gige the average of two number on the same date
```

```python
# we can also get sum, max, min - defoult is mean
df.pivot_table(index="date", columns = "city", margins = True)
```
