```python
#Now we are going to make our date col our index 
#for taht we have to change it into data type
df = pd.read_csv("aapl.csv", parse_dates= ["Date"], index_col="Date")
df.head(5)
```

```python
df.index
```

```python
# we are taking date col as index value because it give us many benifit - like example retrieve jan 2017 stock prices
df.loc["2017-01-03"]
```

```python
df.loc["2017-01"]
```

```python
# for using the range funtion we are going to use sort, then only we can use range inpandas
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df.loc["2017-01-01":"2017-01-07"]
```

```python
# monthly average - 
df.Close.resample('M').mean()
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot()
```
